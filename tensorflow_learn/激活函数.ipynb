```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
x = tf.range(-8, 8, 0.1)
y = tf.nn.relu(x)
plt.plot(x, y, c='red', label='Relu')
plt.legend(loc='best') # 显示左上角标题
plt.grid() # 显示网格线
```
