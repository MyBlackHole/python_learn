```python
list1ss = []
my_dict = {}
for i in root[8]:
    my_d = {}
    print('\n')
    print(i.tag, ':', i.attrib)
    a = []
    g = 0
    for j in i:
        f = 0
        if j.tag != 'contour':
            continue
        else:
            g = g  + 1
            print('j', j.tag, ':', j.attrib)
        for z in j:
            if z.tag == 'pt':
                f = 1 + f
                a.append(z.attrib)
            print(z.tag, ':', z.attrib)
        my_d[g] = f
        a.append({'f':f,'lena': len(a)})
    # list1s.append(f)
    my_dict[i.attrib['name']] = my_d
    list1ss.append(a)
```
